```python
from anthropic import Anthropic
from dotenv import load_dotenv 
```

```python
load_dotenv()
client = Anthropic()
model = "sonnet-4-0"
tokens = 1000
```

```python
def add_user_m(messages, text):
    user_message = {"role" : "user", "content" : text}
    messages.append(user_message)

def add_ast_r(messages, text):
    ast_respond = {"role" : "assistant", "content" : text}
    messages.append(ast_respond)

def chat(messages):
     message = client.messages.create(
        model = model,
        max_tokens = tokens,
        message = messages
     )
     return message.content[0].text  
```

```python
messages = []

while(True):
    user_m = input("> ")
    print("> ", user_m)

    add_user_m(messages, user_m)

    a_rsp =  chat(messages)
    print("> ", a_rsp)
    add_ast_r(messages, a_rsp)
    
```
