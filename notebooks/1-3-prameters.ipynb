````python
from anthropic import Anthropic
from dotenv import load_dotenv
````

````python
load_dotenv()
client = Anthropic()
model = "sonnet-4-0"
tokens = 1000
````

````python
def user_m(messages, text):
    user_m = { "role" : "user", "content" : text}
    messages.append(user_m)

def ast_m(messages, text):
    user_a = { "role" : "assistant", "content" : text}
    messages.append(user_a)

def chat(messages, system = None, temperature = 1.0):
    params = {
        "model" : model,
        "max_tokens" : tokens,
        "messages" : messages,
        "tempurature" : temperature
    }
    if system:
        params["system"] = system

    response = client.messages.create(**params)

    return response.content[0].text 
````

````python
messages = []
system = " your a  patient math tutor.  do not directly answer a student questions. gide them to a solution step by step "
user_m(messages, "How i somde 5x+ 2 = 3 ?")
ast_rspn = chat(messages, system=system, temperature=0.0)

ast_rspn
````

````python
# streaming using SDk to get only the response chunks 

messages = []
user_msg = user_m(messages, "Descripe me the coffe in a sentence ")

with client.messages.stream(
    model= model,
    max_tokens= tokens,
    messages= messages
) as stream:
    for text in stream.text_stream:
        print(text, end = "")

assistant_rsp = stream.get_final_message()
````
