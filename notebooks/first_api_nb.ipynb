```python
# create an Api client
from anthropic import Anthropic 

client = Anthropic()
model = "claude-sonnet-4-0"
tokens = 1000
```

```python
# make a request 
message = client.messages.create(
    model =model,
    max_tokens = tokens,
    messages= [
        {
            "role": "user",
            "content": "what is quatum computing ? Answer in one sentence" 
        }

    ]
)
```

```python
# api response 
message.content[0].text
```

```python
# make a conversation with context 

def add_user_message(messages, text):
    user_message = {"role" : "user", "content": text}
    messages.append(user_message)

def add_assistant_responses(messages, text):
    assistant_response = {"roles" : "assistant", "context" : text }
    messages.append(assistant_response) 

def chat(messages):
    message = client.messages.create(
        model = model,
        max_tokens = tokens,
        messages = messages
    )
    return message.content[0].text
```

```python
# make a starting list of messages 
messages = []

# add in the initial user question
user_message = "what is ai ?"
add_user_message(messages, user_message) 


# pass the list of messages into chat to get an answer 
assistant_response = chat(messages)

# add the response to our list 
add_assistant_responses(messages, assistant_response)

#add user follow up 
user_follow_up = "another perspective "
add_assistant_responses(messages,user_follow_up)


assistant_follow_up = chat(messages)
```
